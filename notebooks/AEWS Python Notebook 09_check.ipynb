# AEWS Python Notebook 09_check: visualising some results

```python
%matplotlib inline
from netCDF4 import Dataset, num2date, date2num
import numpy as np
import xarray as xr
import matplotlib.image as img
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = (10,10)   # increase plot size a bit...
rcParams['axes.formatter.useoffset'] = False   # disable scalar formatter / offset in axes labels
```

```python
# save_dir = '/g/data/jr4/vis_data_v2.0/'
# save_dir = '/g/data/jr4/vis_data_v3.0/'
save_dir = '/g/data/jr4/vis_data_v3.1/'
```

```python
# roi_save_name = save_dir + 'ULAH.nc'
# roi_save_name = save_dir + 'EIGHT_MILE.nc'

# roi_save_name = save_dir + 'TELEPHONE.nc'
roi_save_name = save_dir + 'TRUNKETABELLA.nc'
# roi_save_name = save_dir + 'CORINYA.nc'
# roi_save_name = save_dir + 'HUNGERFORD.nc'
# roi_save_name = save_dir + 'BROADMEADOWS.nc'

roi_nc_grp = Dataset(roi_save_name, mode='a')

dat = roi_nc_grp.variables['WQ_data']
times = roi_nc_grp.variables['time']
lats = roi_nc_grp.variables['lat']
lons = roi_nc_grp.variables['lon']

times = num2date( list(times), times.units, times.calendar )   # datetime.datetime object

WQ_xarray_TS_geo = xr.DataArray( np.array(dat), coords=[ ('time', times), ('lat', lats), ('lon', lons) ] )
WQ_xarray_TS_geo.plot(col='time',col_wrap=5);

print( "Satellite flags are:" )
print( roi_nc_grp.variables['satellite_flags'][:] )

roi_nc_grp.close()
```
